## FILTRADO DE DATOS DE SPOTIFY

```python
# importamos librerías necesarias 

import pandas as pd
import json
import requests
import os # no hay que instalarlo
from dotenv import load_dotenv # esta es una librería para poder encriptar nuestra api_key
```

```python
#aqui seleccionamos los años y generos que queremos:
years = [2000, 2001, 2002, 2003]
genres = ['hip-hop', 'pop', 'latin', 'rap', 'reggaeton']

with open('definitivo.json') as archivo:
    data = json.load(archivo)

canciones_filtrada = []

for cancion in data:
    if cancion['id'] != '0G4oMxDMxyPlz8Vileptq8':
        if cancion['genre'] in genres:
            if cancion['year'] in years:
                canciones_filtrada.append(cancion)

canciones_filtrada
```

```python
len(canciones_filtrada) #para saber cuantas cancines tenemos
```

```python
# para tener las canciones en una tupla
datos_canciones_filtrada = [] 

for cancion in canciones_filtrada:
    if cancion.get('artist_name') != 'Tranceistors':
        tupla_canciones = (
            cancion.get("id", "unknown"),
            cancion.get("artist_name", "unknown"),
            cancion.get("track_name", "unknown"),
            cancion.get("genre", "unknown"),
            cancion.get("year", "unknown"))

    datos_canciones_filtrada.append(tupla_canciones)
datos_canciones_filtrada
```

```python
# creamos un data frame con las canciones de los años y generos que queremos:
df_canciones_filtrada = pd.DataFrame(datos_canciones_filtrada, columns = ['id_cancion', 'Nombre_artista', 'Nombre_cancion', 'Género', 'Año'])
df_canciones_filtrada
```

```python
df_canciones_filtrada[df_canciones_filtrada["Año"]==2003]["Nombre_cancion"].unique()
```

```python
artistas = []

for cancion in datos_canciones_filtrada:
    nombre_artista = cancion[1]
    
    if nombre_artista not in artistas:
        artistas.append(nombre_artista)

artistas
```

```python
len(artistas)
```

### Extracción datos con ID cancion, nombre_cancion, año y ID artista

```python
artistas_dict = {}      # Guarda artista -> id_autoincremental
canciones = []         # Lista final con canciones
id_artista = 1         # Contador autoincremental

for cancion in datos_canciones_filtrada:
    nombre_artista = cancion[1]
    
    # Si el artista no existe, se le asigna nuevo ID
    if nombre_artista not in artistas_dict:
        artistas_dict[nombre_artista] = id_artista
        id_artista += 1
    
    # Crear registro de canción con ID del artista
    registro = {
        "spotify_id": cancion[0],
        "artista_id": artistas_dict[nombre_artista],
        "artista": nombre_artista,
        "cancion": cancion[2],
        "genero": cancion[3],
        "anio": cancion[4]
    }
    
    canciones.append(registro)
```

### Asignar ID género de cada género

```python
def generar_id_genero(canciones):
    # Paso 1: obtener géneros únicos
    generos_unicos = sorted(set([c[3] for c in canciones]))  # c[3] es el género
    # Paso 2: crear un diccionario género -> id
    genero_id_map = {genero: idx+1 for idx, genero in enumerate(generos_unicos)}
    return genero_id_map
```

```python
generar_id_genero(datos_canciones_filtrada)
```

### Tabla canciones (columnas: spotify_id & cancion & año & artista_id)

```python
tabla_canciones = [] 

for cancion in canciones:
    fila = (cancion['spotify_id'], 
            cancion['cancion'],
            cancion['anio'],
            cancion['artista_id'],
            ) # Cada genero con su ID

    tabla_canciones.append(fila)
tabla_canciones
```

```python
len(tabla_canciones)
```

### Tabla intermedia: ID canción, ID género

```python
tabla_intermedia = [] 

for cancion in canciones:
    nombre_genero = cancion['genero']
    if nombre_genero == "hip-hop":
        id_genero = 1
    elif nombre_genero == "latin":
        id_genero = 2
    elif nombre_genero == "pop":
        id_genero = 3
    elif nombre_genero == "regaetton":
        id_genero = 4
    elif nombre_genero == "rap":
        id_genero = 5 

    fila = (cancion['spotify_id'], 
            id_genero
            ) # Cada genero con su ID

    tabla_intermedia.append(fila)
tabla_intermedia
```
